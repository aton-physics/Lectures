## problem 1a
The adiabatic model is very important in determining how far a launched ball may go. Since the drag force in any direction depends not only on the velocity in that direction, but also the magnitude of the total velocity, the launch angle also becomes very important. A large vertical velocity will incur a large drag force altering both the vertical and horizontal velocities. So a very large (near-vertical) angle is very terrible if the goal is to get a large horizontal distance before landing. Air is thinner in the higher altitudes, so there is less overall drag force; the ball goes farther. In the constant density approximation, the trajectory does not go as far since the drag force is larger. Finally, the ball in free fall goes furthest as expected, as the drag force is totally absent.

## problem 1b

The mesosphere begins at about 50,000 meters above sea level. So we will use 60,000 meters as our initial altitude. Further, the adiabatic model breaks down at altitudes higher than 45,000 meters so we take the air density to be zero past this height. Also, the gravitational acceleration due to height can be given as $$a = GM/(R+h)^2$$

where the variables take their usual definitions in Newtonian mechanics: a is acceleration, G is gravitational constant, M is the mass of the earth, R is the radius of the earth to the surface, and h is the altitude of our particle.

```python
constantdensityball = Particle(0, 100, 50000, 10, 10, 0.01, "Constant air density")
constantdensityball.scipy_trajectory(1)
constantdensityball.plot()

freeball = Particle(0, 100, 50000, 10, 10, 0.01, "No drag: free ball")
freeball.scipy_trajectory(2)
freeball.plot()
```

Clearly, gravitational acceleration's variation with height does not severely alter the particle's trajectory. Consider the equation for gravitational acceleration and this may be obvious: the earth is long relative to the distance between the surface and the mesosphere. R is about 6371 km, while the mesosphere sits at 50 km, so the perturbation to the height is one part in a hundred and can be neglected to a good approximation as we have shown.

## Problem 2d

It is probably not fair to assume constant angular velocity of our basketball. Note that if the (motion-opposing) torque is proportional to the angular velocity we have $$ \theta'' = -\theta'$$ which is solved by decaying exponentials (with a time constant and initial condition).

I will explore the simplest exponential $$ \theta'(t) = \theta'_{o}e^{-t}$$

```python
#damball = Particle3D(0, 0, 0, 0, 100, 1, 100, 0.01, "3d plot, thrown straight up with backspin off a dam (no wind)", 0, 4*3.14, 0)
damball = Particle3D(0, 0, 0, 0, 100, 1, 33, 0.01, "3d plot, thrown straight up with backspin off a dam (no wind)", 0, 4*3.14, 0)
damball.RK4_trajectory()
damball.RK4_plot()
#x0=0.0, v_x0=0.0,  y0=0.0, v_y0 = 0.0, z = 100.0, v_z0 = 0.0, tf = 10.0, dt = 0.01, tag = "tag", omegax = 0.0, omegay = 0.0, omegaz = 0.0
```

As expected, an exponentially decaying rotational velocity means that the magnus force must also decay. I.e. the ball will lose its lift, and drag will make sure it plummets after some time. The ball is again thrown straight up with backspin, but this time it does not go nearly as far (less than 2 meters).

Conclusion: The trajectory is very different for an exponentially decaying rotational velocity, so it is likely not fair to assume it is constant.
